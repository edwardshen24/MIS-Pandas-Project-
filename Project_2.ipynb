```python
import pandas as pd
xl=pd.ExcelFile("/content/TableauSalesData.xlsx")
SalesData = xl.parse("Orders")
```

```python

#Weâ€™d like to drop underperforming products or even subcategories, but we still want to provide all the products requested by top 50% most frequent and profitable accounts.
#https://docs.google.com/document/d/1vqH5oyNJUod03FhUCtqbULR4-PdIy6GP/edit
JustTables = SalesData.loc[SalesData["Sub-Category"]=="Tables"]
JustBookcases = SalesData.loc[SalesData["Sub-Category"]=="Bookcases"]
JustSupplies = SalesData.loc[SalesData["Sub-Category"]=="Supplies"]
Totals = pd.concat([JustTables, JustBookcases, JustSupplies])
InventoryQuantity = SalesData[["Product Name","Sales","Quantity"]]

#edward
def LeastProfitableInSubcategory():
  print(Totals[["Product Name", "Profit", "Sales"]].groupby("Product Name").sum().sort_values(by="Profit").head(5))

def MostProfitableInSubcategory():
  print(Totals[["Product Name", "Profit", "Sales"]].groupby("Product Name").sum().sort_values(by="Profit",ascending=False).head(5))


def MostProfitableOverall():
  print(SalesData[["Product Name", "Profit","Segment"]].groupby("Product Name").sum().sort_values(by="Profit",ascending=False).head(10))

def MostProfitableBySegment():
  Consumer = SalesData.loc[SalesData["Segment"]=="Consumer"]
  Corporate = SalesData.loc[SalesData["Segment"]=="Corporate"]
  HomeOffice = SalesData.loc[SalesData["Segment"]=="Home Office"]
  print("\n Consumer Segment")
  print(Consumer[["Product Name", "Profit", "Sales"]].groupby("Product Name").sum().round().sort_values(by="Profit", ascending = False).head(5))
  print("\n Corporate Segment")
  print(Corporate[["Product Name", "Profit", "Sales"]].groupby("Product Name").sum().round().sort_values(by="Profit", ascending = False).head(5))
  print("\n Home Office Segment")
  print(HomeOffice[["Product Name", "Profit", "Sales"]].groupby("Product Name").sum().round().sort_values(by="Profit", ascending = False).head(5))

def MostQuantity():
  print(InventoryQuantity.groupby(by=["Product Name"]).sum().sort_values(by = ['Sales', 'Quantity'], ascending = [True, False]).head(10))

def LeastQuantity():
  print(InventoryQuantity.groupby(by=["Product Name"]).sum().sort_values(by = ['Sales', 'Quantity'], ascending = [True,True]).head(10))

#zachary
def MostprofitableCustomers():
  Customerdata=SalesData[["Customer ID","Customer Name","Product Name","Profit"]]
  Customerprofit=Customerdata.groupby(by=['Customer ID']).sum().sort_values(by='Profit',ascending=False).head(10)
  print(Customerprofit)

def TopCustomerProducts():
  Customerdata=SalesData[["Customer ID","Customer Name","Product Name","Profit"]]
  Customerprofit=Customerdata.groupby(by=['Customer ID']).sum().sort_values(by='Profit',ascending=False).head(10)
  Customer=SalesData[['Customer ID','Product Name']]
  topcustomers = ['TC-20980','RB-19360','SC-20095','HL-15040','AB-10105','TA-21385','CM-12385','KD-16495','AR-10540','DR-12940']
  for customers in topcustomers:
    Customerproducts=Customer.loc[Customer['Customer ID']==customers]
    print('Customer '+customers+' purchased the following:')
    print(Customerproducts)
    print('-'*40)

def Bottomsubcategories():
  BotSubCats = SalesData[['Sub-Category','Profit','Sales']]
  BotTotalSubCats = BotSubCats.groupby(by = 'Sub-Category').sum().sort_values(by = 'Profit').head(5)
  print(BotTotalSubCats)

def Botproductssubcats():
  Botsubcats = ['Tables','Bookcases','Supplies','Fasteners','Machines']
  Data = SalesData[['Sub-Category','Product Name','Profit']]
  for cats in Botsubcats:
    Catproducts=Data.loc[Data['Sub-Category']==cats].sort_values(by='Profit',ascending = True).head(5)
    print('The bottom 5 products by profit in the '+cats+' Sub-Category are the following:')
    display(Catproducts)
    print('-'*40)

def Yearlybotproducts():
  sales = SalesData
  salesyears = sales.copy()
  salesyears['Year'] = salesyears['Order Date'].dt.year
  Years = salesyears.Year.unique()
  salesbyyear = salesyears[['Product Name','Year','Profit',]]
  for year in Years:
    OneYear = salesbyyear.loc[salesbyyear['Year']==year]
    NoYear = OneYear[['Product Name','Profit']]
    YearlyRegSales = NoYear.groupby(by='Product Name').sum().sort_values(by='Profit').head(10)
    print('Bottom 10 performing products for the year: ' + str(year))
    print(YearlyRegSales)
    print('*'*40)

def Yearlysubcats():
  sales = SalesData
  salesyears = sales.copy()
  salesyears['Year'] = salesyears['Order Date'].dt.year
  Years = salesyears.Year.unique()
  salesbyyear = salesyears[['Sub-Category','Year','Profit',]]
  for year in Years:
    OneYear = salesbyyear.loc[salesbyyear['Year']==year]
    NoYear = OneYear[['Sub-Category','Profit']]
    YearlyRegSales = NoYear.groupby(by='Sub-Category').sum().sort_values(by='Profit').head(5)
    print('Bottom 5 performing sub-categories for the year: ' + str(year))
    print(YearlyRegSales)
    print('*'*40)

#main menu 
print("Welcome to the Office Solutions Data Analytics System! To learn more about our profits and sales,"+ 
      "\n underperforming and profitable products, and quantity of items in inventory enter a number below.") 
def menu():
  print("\n Enter 1 to see the 5 least profitable products across all sub-categories"+
        "\n Enter 2 to see the 10 most profitable products overall" +
        "\n Enter 3 to see the 10 most amount of items in inventory by sales"+
        "\n Enter 4 to see the 10 most profitable customer accounts"+
        "\n Enter 5 to see the 5 least profitable sub-categories"+
        "\n Enter 6 to see the 10 least profitable products by year"+
        "\n Enter 7 to exit")
  choice = input("\n Please enter your selection here: ")
  print("\n")
  #edward
  if choice == "1":
    LeastProfitableInSubcategory()
    print("\n Enter 1 to see the 5 most profitable products in the least profitable sub-categories"+
          "\n Enter anything else to go back")
    choice = input("\n Please enter your selection here: ")
    if choice == "1":
      MostProfitableInSubcategory()
      menu()
    else:
      menu()

  elif choice == "2":
    MostProfitableOverall()
    print("\n Enter 1 to see the 5 most profitable products broken down by customer segment"+
          "\n Enter anything else to go back")
    choice = input("\n Please enter your selection here: ")
    if choice == "1":
      MostProfitableBySegment()
      menu()
    else:
      menu()

  elif choice == "3":
    MostQuantity()
    print("\n Enter 1 to see the 10 least amount of items in inventory by sales"+
          "\n Enter anything else to go back")
    choice = input("\n Please enter your selection here: ")
    if choice == "1":
      LeastQuantity()
      menu()
    else:
      menu()

  #zachary
  elif choice == '4':
    MostprofitableCustomers()
    print('\n Enter 1 to see the products purchased by the 10 most profitable accounts'+
          '\n Enter anything else to go back')
    number = input('\n Please enter your selection here')
    if number == '1':
      TopCustomerProducts()
      menu()
    else:
      menu()
    
  elif choice == "5":
    Bottomsubcategories()
    print('\n Enter 1 to see the least profitable product in these 5 sub-categories')
    print('Enter anything else to go back')
    number = input('Please enter your selection here')
    if number == '1':
      Botproductssubcats()
      menu()
    else:
      menu()

  elif choice == "6":
    Yearlybotproducts()
    print('\n Enter 1 to see the 5 least profitable segments by year')
    print('Enter anything else to go back')
    number = input('Please enter your input here')
    if number =='1':
      Yearlysubcats()
      menu()
    else:
      menu()

  elif choice =="7":
    quit()

  else:
    print("You have entered an invalid option, please select from one of the options provided: ")
    menu()

menu()
```
